```python
class Node:
     
    def __init__(self,Share):
        self.Share=Share
        self.Next=None
            
class LinkedList:
    
    def __init__(self):
        self.head=None
        
        
    def insertShare(self,newNode):
        if self.head== None:
            self.head=Node(newNode)
        else:
            currentNode=self.head
            while True:
                if currentNode.Next == None:
                    break
                currentNode.Next=currentNode.Next
            currentNode.Next=newNode
        
    def deleteNode(self, key): 
          
        # Store head node 
        temp = self.head 
  
        # If head node itself holds the key to be deleted 
        if (temp is not None): 
            if (temp.Share == key): 
                self.head = temp.next
                temp = None
                return
  
        # Search for the key to be deleted, keep track of the 
        # previous node as we need to change 'prev.next' 
        while(temp is not None): 
            if temp.Share == key: 
                break 
            prev = temp 
            temp = temp.Next 
  
        # if key was not present in linked list 
        if(temp == None): 
            return 
  
        # Unlink the node from linked list 
        prev.Next = temp.Next 
  
        temp = None
  
  
    # Utility function to print the linked LinkedList 
    def printList(self): 
        if self.head==None:
            print("list is empty")
        else:
            lastnode=self.head
            while True:
                if lastnode==None:
                    break
                print(lastnode.Share)
                lastnode=lastnode.Next
        
  

#    if __init__== "__main__":
#        for index in range(5):
#            llist = LinkedList() 
#            Share=int(input("please enter the value of share"))
#            llist.push(Share)
#           llist.printList()
            
llist = LinkedList() 
llist.insertShare(('ee',0)) 
#llist.insertShare(1) 
#llist.insertShare(3)
#llist.insertShare(2) 
  
print("Share List: ")
llist.printList() 
llist.deleteNode(1)  
print("\nLinked List after Deletion of 1:")
llist.printList() 
    
```
